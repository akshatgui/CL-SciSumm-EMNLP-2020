```python
def clean(line,root,final1,final2):
    line = line.split("|")
    cit_2 , ref_2 = line[6].split("Citation Text:")[1].strip() , line[8].split("Reference Text:")[1].strip()
    cit_2 = cit_2.split(" >")[1].split("</S>")[0]
    ref_2 = ref_2.split(" >")[1:]
    ref_2 = [ i.split("</S>")[0] for i in ref_2]
    #print(ref_2)
    
    ans=cit_2
    for a in root:
        for b in a:
            flag=0
            final1.append(cit_2)
            final2.append(b.text)
            for i in ref_2:
                if i in b.text:
                    out.append("1")
                    flag=1
                    break
                
            
            if flag==0:
                out.append("0")
    
    
    return ans
```

```python
import xml.etree.ElementTree as ET
tree = ET.parse("./Reference_XML/W02-1503.xml")
root = tree.getroot()
out=[]
final1=[]
final2=[]

with open("./annotation/W02-1503.ann.txt") as f:
    data=f.read()
    arr=[]
    data = [ clean(i,root,final1,final2) for i in data.split("\n") if i ]
    
    
```
