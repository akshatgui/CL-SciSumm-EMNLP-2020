```python
import xml.etree.ElementTree as ET
import random
tree = ET.parse('A00-1031.xml')
```

```python
root = tree.getroot()

#root = ET.fromstring(A00-1031_as_string)
```

```python
root.tag
```

```python
for child in root:
    for i in child:
        print(i.text)
```

```python
i=0
j=0
while i < len(root):
    j=0
    while j<len(root[i]):
        print(root[i][j].attrib)
        print(root[i][j].text)
        j=j+1
    i=i+1
```

```python
attrib=len(root)
rand0=random.randint(0,attrib-1)
lent=len(root[rand0])
if lent>1:
    rand1=random.randint(0,lent-1)
try:
    print(root[rand0][rand1].text)
except:
    print(1)
```
